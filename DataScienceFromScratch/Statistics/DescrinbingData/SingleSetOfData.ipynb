```python
from collections import Counter
import matplotlib.pyplot as plt
```

```python
num_friends = [100,49,41,40,25,79,10,5,25,33,51,58,69,88]
friend_counts = Counter(num_friends)

xs = range(101)  #largest value is 100
ys = [friend_counts[x] for x in xs] #height is just # of friends
plt.bar(xs,ys)
plt.axis([0,101,0,25])
plt.title('Histogram of Friends Counts')
plt.xlabel('# of Friends')
plt.ylabel('# of People')
plt.show()
```

```python
num_point = len(num_friends)  #number of data points/amount of data
largerst_value = max(num_friends) # maximum value
smallest_value = min(num_friends)# minimum value
sorted_values = sorted(num_friends)
second_smallest_value = sorted_values [1] # second maximum value
second_largers_value = sorted_values [-2] # second minimum value
print(f'Number of occurrences {num_point}')   
print(f'Biggest value  {largerst_value}')  
print(f'Second biggest value  {second_largers_value}')    
print(f'Minimum value  {smallest_value}')   
print(f'Second minimum value  {second_smallest_value}')   
```
